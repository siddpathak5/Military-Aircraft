```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_active_Indian_military_aircraft'
```

```python
headers = {
    'User-Agent': 'Brave (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/41.0.2228.0 Safari/537.36'
}
```

```python
response = requests.get(url,headers)
```

```python
response
```

```python
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
soup
```

```python
table = soup.find('table', class_='wikitable')
```

```python
table
```

```python
headers = []
for i in table.find_all('th'):
    headers.append(i.text.strip())
```

```python
headers
```

```python
table.find_all('tr')
```

```python
for i in table.find_all('tr')[2:]:
    print(i)
    print("----")
```

```python
aircraft_type=[]

for i in table.find_all('tr')[2:]:
    aircraft_type.append(i.find('a').text)
```

```python
aircraft_type
```

```python
for i in table.find_all('td')[2:]:
    print(i)
    print("----")
```

```python
origin = []

for td in table.find_all('td'):
    a_tag = td.find('a')
    if a_tag:
        origin.append(a_tag.text)
```

```python
origin
```

```python
df1 = pd.DataFrame(aircraft_type)
```

```python
df1.to_csv('aircraft',index=False)
```
